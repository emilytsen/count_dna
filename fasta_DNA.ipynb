```python
def ler_fasta(arquivo):
  sequencia = ''
  lista = []
  with open(arquivo, 'r') as fasta:
    for linha in fasta:
      sequencia = ''
      if not linha.startswith('>'):
          sequencia += linha
      lista.append(sequencia)

  return lista

lista = ler_fasta('/content/data/sequence.fasta')

for i in lista:
  print(i)
```

```python
def contar_nucleotideos(lista):
    contagens = {'C': 0, 'A': 0, 'T': 0, 'G': 0}
    total_nucleotideos = 0
    for sequencia in lista:
        for nucleotideo in sequencia:
            if nucleotideo in contagens:
                contagens[nucleotideo] += 1
                total_nucleotideos += 1
    return contagens, total_nucleotideos

def calcular_porcentagens(contagens, total_nucleotideos):
    porcentagens = {}
    for nucleotideo, contagem in contagens.items():
        porcentagens[nucleotideo] = (contagem / total_nucleotideos) * 100
    return porcentagens

contagens, total_nucleotideos = contar_nucleotideos(lista)
porcentagens = calcular_porcentagens(contagens, total_nucleotideos)

print("Contagem de nucleotídeos:")
for nucleotideo, contagem in contagens.items():
    print(f"{nucleotideo}: {contagem}")

print("\nPorcentagem de cada nucleotídeo:")
for nucleotideo, porcentagem in porcentagens.items():
    print(f"{nucleotideo}: {porcentagem:.2f}%")

print(f"\nTotal de nucleotídeos: {total_nucleotideos}")
```
